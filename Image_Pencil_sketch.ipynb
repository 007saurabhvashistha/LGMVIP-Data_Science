```python
from PIL import Image
image_path = '/content/drive/My Drive/DOG.JPG'
image = Image.open(image_path)
image.show()
```

```python

# Convert the PIL Image to a NumPy array
image_np = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

#Reading the black and white image

b_and_w_image = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)

#Displaying the Black and White Image
cv2_imshow(b_and_w_image)
cv2.waitKey(0)
```

```python
#iInverting the image

inverted_image = 255 - b_and_w_image

#Displaying the Inverted Image
cv2_imshow(inverted_image)
cv2.waitKey()
```

```python
#blurring the image
blurred = cv2.GaussianBlur(inverted_image,(21,21),0)
```

```python
#Inverting the Blurred image

inverted_blurred =255- blurred

# Creating the Pencil Sketch image

pencil_skecth =cv2.divide(b_and_w_image,inverted_blurred,scale=256.0)
#displaying the Skecth Image
cv2_imshow(pencil_skecth)
cv2.waitKey(0)
```

```python
# Displaying Original Image
cv2_imshow(image_np)
# Displaying Sketch Image
cv2_imshow(pencil_skecth)
cv2.waitKey(0)
```
